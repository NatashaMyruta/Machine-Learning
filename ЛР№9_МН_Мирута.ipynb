```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
from matplotlib.lines import Line2D

# Обробка даних та декомпозиція
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

# Кластеризація
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from kneed import KneeLocator
from yellowbrick.cluster import KElbowVisualizer

# Класифікація та метрики
from sklearn.ensemble import GradientBoostingClassifier, HistGradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.inspection import permutation_importance
```

```python
path = kagglehub.dataset_download("brunogrisci/leukemia-gene-expression-cumida")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(os.path.join(path, "Leukemia_GSE9476.csv"))
df
```

```python
target = df["type"]
target.value_counts()
```

```python
X = df.drop(columns=['samples', 'type'])
```

```python
X = StandardScaler().fit_transform(X)

pca = PCA(random_state=42).fit(X)
pve = pca.explained_variance_ratio_
```

```python
sns.set_theme()

kneedle = KneeLocator(
    x=range(1, len(pve) + 1),
    y=pve,
    curve='convex',
    direction='decreasing'
)

# Візуалізація
kneedle.plot_knee()
plt.title(f'Knee Point at {kneedle.elbow + 1}')
plt.xlabel("Number of components")
plt.ylabel("Explained Variance Ratio")
plt.grid(True)
plt.show()
```

```python
n_components = kneedle.elbow

ax = sns.lineplot(np.cumsum(pve))

ax.axvline(x=n_components,
           c='black',
           linestyle='--',
           linewidth=0.75)

ax.axhline(y=np.cumsum(pve)[n_components],
           c='black',
           linestyle='--',
           linewidth=0.75)

ax.set(xlabel='number of components',
       ylabel='cumulative explained variance')

plt.show()
```

```python
X = pca.transform(X)[:, :n_components]
```

```python
model_kmn = KMeans(random_state=42)

visualizer = KElbowVisualizer(
    model_kmn,
    k=(2, 10),
    timings=False)

visualizer.fit(X)
visualizer.show()
```

```python
def plot_dendrogram(model, **kwargs):

    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)

    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1 # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    dendrogram(linkage_matrix, **kwargs)
```

```python
model_agg = AgglomerativeClustering(
    distance_threshold=0,
    n_clusters=None)

model_agg = model_agg.fit(X)
```

```python
plot_dendrogram(model_agg, truncate_mode='level', p=3)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Number of points in node')

plt.show()
```

```python
k_best = visualizer.elbow_value_

model_kmn = KMeans(n_clusters=k_best, random_state=42).fit(X)
model_agg = AgglomerativeClustering(n_clusters=k_best).fit(X)

labels_kmn = pd.Series(model_kmn.labels_, name='k-means')
labels_agg = pd.Series(model_agg.labels_, name='h-clust')
```

```python
pd.crosstab(df['type'], labels_kmn)
```

```python
pd.crosstab(df['type'], labels_agg)
```

```python
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(10, 5), sharey=True)

for i, s in enumerate([df['type'], labels_kmn]):
    ax = axes[i]

    sns.scatterplot(x=X[:, 0],
                    y=X[:, 1],
                    hue=s,
                    style=s,
                    edgecolor='black',
                    linewidth=0.5,
                    s=60,
                    palette='tab20',
                    legend=False,
                    ax=ax)

    ax.set(title=s.name)

plt.show()
```

```python
from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import kagglehub
from scipy.stats import zscore
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import silhouette_score
from matplotlib.lines import Line2D
from sklearn.decomposition import PCA
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.inspection import permutation_importance
import os
```

```python
path = kagglehub.dataset_download("brunogrisci/leukemia-gene-expression-cumida")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(os.path.join(path, "Leukemia_GSE9476.csv"))
df
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
target = df["type"]
target.value_counts()
```

```python
data = df.drop(columns=["samples", "type"])

z_scores = data.apply(zscore).abs()

outliers_mask = z_scores > 3

outlier_fraction = outliers_mask.mean(axis=1)

outlier_indices = outlier_fraction[outlier_fraction > 0.2].index

df = df.drop(index=outlier_indices)
target = target.drop(index=outlier_indices)
```

```python
df.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data,
    target,
    test_size=0.2,
    random_state=42
)
```

```python
scaler = StandardScaler().set_output(transform='pandas')

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
pca = PCA().set_output(transform='pandas').fit(X_train)
```

```python
explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Створення графіка
sns.set_theme()
fig, ax = plt.subplots(figsize=(8, 5))

sns.lineplot(x=np.arange(1, len(explained_variance) + 1), y=explained_variance, ax=ax)

ax.set(
    xlabel='Кількість компонент',
    ylabel='Накопичена пояснена дисперсія',
    title='PCA: Кількість компонент для 85% дисперсії'
)

# Знаходимо кількість компонент, достатню для 85%
n_components = np.searchsorted(explained_variance, 0.85) + 1  # +1 бо індексація з 0

# Вертикальна і горизонтальна лінії
ax.axvline(x=n_components, color='black', linestyle='--', linewidth=0.75)
ax.axhline(y=explained_variance[n_components - 1], color='black', linestyle='--', linewidth=0.75)

plt.tight_layout()
plt.show()
```

```python
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
X_train_pca.iloc[:, :n_components].head()
```

```python
# Перетворимо PCA-результати у DataFrame
pca_df = pd.DataFrame(
    X_train_pca,
    columns=[f"pca{i}" for i in range(X_train_pca.shape[1])],
    index=y_train.index # індекс важливо зберегти
)

# Закодуємо текстові мітки класів у числа
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)

# Побудуємо 3D-графік
plt.figure(figsize=(8, 8))
ax = plt.subplot(projection='3d')

scatter = ax.scatter3D(
    pca_df["pca0"],
    pca_df["pca1"],
    pca_df["pca2"],
    c=y_train_encoded,
    cmap='autumn',
    s=20,
    edgecolor='black',
    linewidth=0.75
)

# Оформлення графіка
ax.set_ylabel("PCA 1")
ax.set_zlabel("PCA 2")
ax.view_init(elev=30, azim=30)

# Кольори для кожного класу
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label=label,
           markerfacecolor=plt.cm.autumn(i / len(label_encoder.classes_)),
           markeredgecolor='black', markersize=8)
    for i, label in enumerate(label_encoder.classes_)
]

ax.legend(handles=legend_elements, title="Класи", loc='best')

# 6. Показуємо
plt.tight_layout()
plt.show()
```

```python
clf_full = HistGradientBoostingClassifier(max_iter=50, random_state=42)
clf_full.fit(X_train, y_train)
```

```python
clf_pca = GradientBoostingClassifier()

clf_pca.fit(X_train_pca.iloc[:, :n_components], y_train)

pred_pca = clf_pca.predict(X_test_pca.iloc[:, :n_components])

score_pca = accuracy_score(y_test, pred_pca)

print(f'Model accuracy (PCA): {score_pca:.1%}')
```

```python
import shap
import numpy as np

explainer = shap.TreeExplainer(clf_full)
shap_values = explainer.shap_values(X_train)

if shap_values.ndim == 3:
    mean_shap = np.abs(shap_values).mean(axis=(0, 2))
elif shap_values.ndim == 2:
    mean_shap = np.abs(shap_values).mean(axis=0)

importances = pd.Series(data=mean_shap, index=X_train.columns)

top_features = importances.sort_values(ascending=False).head(15)

plt.figure(figsize=(6, 10))
top_features.sort_values().plot.barh(color='teal')
plt.title("Топ-15 важливих ознак (SHAP)")
plt.xlabel("Mean |SHAP value|")
plt.tight_layout()
plt.show()
```

```python
nonzero = importances[importances > 0]
top = nonzero.sort_values(ascending=False).head(30)

plt.figure(figsize=(6, 10))
top.sort_values().plot.barh()
plt.title("Топ важливих ненульових ознак")
plt.tight_layout()
plt.show()
```
